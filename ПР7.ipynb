```python
#8.1
def vse_del(n):
  for i in range(1, n+1):
    t = i
    while t != 0:
      d = t % 10
      if d == 0 or i % d != 0:
          break
      t //= 10
    else:
        print(i, end=' ')
print(vse_del(15))
```

```python
#8.2
def Zamena(Sp):
    if len(Sp) > 1:
        Sp[0], Sp[-1] = Sp[-1], Sp[0]
m = int(input("Введите длину массива: "))
A = []
print("Введите элементы массива:")
for i in range(m):
    element = int(input(f"Элемент {i+1}: "))
    A.append(element)
print("Исходный массив:", A)
Zamena(A)
print("Массив после замены первого и последнего элементов:", A)
```
