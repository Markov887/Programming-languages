2 Задана квадратная матрица. Получить транспонированную матрицу
(перевернутую относительно главной диагонали) и вывести на экран.

```python
from numpy import *
n = int(input("Введите размер матрицы: "))
k = int(input("Введите номер строки k (от 0 до n-1): "))
a = np.random.randint(1, 50, size=(n, n))
print(a)
for i in range(len(a)):
  deli = a[k][i] / a[i][i]
  print(a[k][i], '/', a[i][i], '=', deli)
```
