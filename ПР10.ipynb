1 Задана матрица порядка n и число к. Разделить элементы k-й строки на диагональный элемент, расположенный в этой строке.

2 Задана квадратная матрица. Получить транспонированную матрицу (перевернутую относительно главной диагонали) и вывести на экран.

```python
with open('/Марков_УБ-51_vvod.txt', 'r') as file_vvod:
   n, k = map(int, file_vvod.readline().split())
   matrix = []
   for i in range(n):
      chast = list(map(int, file_vvod.readline().split()))
      matrix.append(chast)
with open('/Марков_УБ-51_vivod.txt', 'w') as file_vivod:
  for i in range(len(matrix)):
    deli = str(matrix[k][i] / matrix[i][i])
    file_vivod.write(deli + '\n')
```

```python
from numpy import *
with open('/Марков_УБ-51_vvod2.txt', 'r') as file_vvod2:
   n, k = map(int, file_vvod2.readline().split())
   matrix = []
   for i in range(n):
      chast = list(map(int, file_vvod2.readline().split()))
      matrix.append(chast)
transposed = []
for i in range(n):
    chast1 = []
    for j in range(n):
        chast1.append(matrix[j][i])
    transposed.append(chast1)
with open('/Марков_УБ-51_vivod2.txt', 'w') as file_vivod2:
  for z in range(len(matrix)):
    file_vivod2.write(str(matrix[z]) + '\n')
  file_vivod2.write('\n')
  for l in range(len(transposed)):
    file_vivod2.write(str(transposed[l]) + '\n')
```
